```python
from sklearn import set_config

set_config(transform_output="pandas")
```

```python
import pandas as pd
import numpy as np
train_data = pd.DataFrame({
    'Fruit': ['Apple',"Apple", 'Banana', "Banana",'Cherry']
})
test_data = pd.DataFrame({
    'Fruit': ['Apple',"Apple", 'Banana', "Banana", 'Grape',"Strawberry", np.nan]
})
```

```python
from category_encoders import BinaryEncoder

encoder = BinaryEncoder(cols=['Fruit'], handle_unknown='indicator', handle_missing="indicator")
encoder.fit(train_data)

encoder.transform(train_data)
```

```python
from IPython.display import display

display(test_data, encoder.transform(test_data))

```
