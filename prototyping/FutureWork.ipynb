```python
import pandas as pd
from pyod.models.auto_encoder import AutoEncoder
from sklearn import set_config

# set_config(transform_output="pandas")
```

```python
df = pd.DataFrame({
    "Kategorie":["Katze","Katze","Katze","Katze","Katze","Katze","Maus",
                 "Hund","Hund","Hund","Hund","Hund","Hund","Hund","Giraffe"]
})
df
```

```python
from category_encoders import BinaryEncoder

encoder = BinaryEncoder(handle_unknown='indicator', handle_missing="indicator")
encoder.fit(df)

transformed_df = encoder.transform(df)
transformed_df
```

```python
clf = AutoEncoder(contamination=0.1, hidden_neurons=[3,1,3], epochs=20)
clf_fitted = clf.fit(transformed_df)
pred_error = clf_fitted.decision_function(transformed_df)
# pred_error = clf_fitted.predict(transformed_df)
```

```python
df["pred_error"] = pred_error
df
```
